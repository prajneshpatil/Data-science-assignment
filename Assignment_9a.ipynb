## Apriori Algorithm

```python
### with min support=0.1
```

```python
frequent_books = apriori(df, min_support=0.1, use_colnames=True)
frequent_books
```

```python
### with minimum support criterion of 0.1 frequent set generated is 39
```

```python
rules = association_rules(frequent_books, metric="lift", min_threshold=0.7)
rules
```

```python
### with lift ratio =0.7 on min support criterion of 0.1, 100 frequent sets are generated
```

```python
rules = association_rules(frequent_books, metric="lift", min_threshold=1.9)
rules
```

```python
frequent_books = apriori(df, min_support=0.2, use_colnames=True)
frequent_books
```
